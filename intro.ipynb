## Preprocessing : Mobility

Creates `dfMOB`, `dfMOB_state`, `dfMOB_county`

```python
dfMOB.head()
```

```python
dfMOB = dfMOB.rename(columns={'retail_and_recreation_percent_change_from_baseline':'retail',
                  'grocery_and_pharmacy_percent_change_from_baseline':'grocery_pharmacy',
                  'transit_stations_percent_change_from_baseline':'transit',
                  'workplaces_percent_change_from_baseline':'workplaces',
                  'residential_percent_change_from_baseline':'residential',
                 'parks_percent_change_from_baseline':'residential'
                  })
```

```python
dfMOB = dfMOB.dropna(subset=['sub_region_1'])
```

```python
dfMOB
```

```python
ogColumns = dfMOB.columns.tolist()

for removeName in ['country_region_code',
               'country_region',
               'metro_area',
               'sub_region_1',
              'census_fips_code',
              'place_id',]:
    ogColumns.remove(removeName)
    
dfMOB = dfMOB[ogColumns]
dfMOB_state = dfMOB.dropna(subset=['iso_3166_2_code'])
dfMOB_county = dfMOB.dropna(subset=['sub_region_2'])
dfMOB_state['iso_3166_2_code'] = dfMOB_state['iso_3166_2_code']\
                                .apply(lambda x: x.split('-')[1])
```

## Preprocessing : Policy 

creates `NATPOLCHANGE` and `STATEPOLCHANGE`

### National Policy

```python
# National 
dfPOL[dfPOL['Jurisdiction']=='NAT_TOTAL']
# State Policies
dfPOL[dfPOL['Jurisdiction']=='STATE_TOTAL']
```

```python
# Policy Data 
dfPOLnat = dfPOL[dfPOL['Jurisdiction']=='NAT_TOTAL']
```

```python
polColumns = [
    'C1_School closing',
    'C6_Stay at home requirements',
    'C2_Workplace closing',
    'C7_Restrictions on internal movement',
    'C3_Cancel public events',
    'C4_Restrictions on gatherings',
    'C5_Close public transport',
    'H2_Testing policy'
]
```

```python
# National Policies
dfPOLnat = dfPOLnat[polColumns+['Date']]
```

```python
dfPOLnat = dfPOLnat.sort_values(by=['Date'])
```

```python
for policyName  in polColumns:
    dfPOLnat['diff_'+policyName] = dfPOLnat.loc[:,policyName]\
    .rolling(window=2)\
    .apply(lambda x: x.iloc[1] - x.iloc[0])
```

```python
dfPOLnat
```

```python
DATELIST = []

for policyName  in polColumns:
    diffColumn = 'diff_'+policyName
    
    dates_of_change = dfPOLnat.dropna()[dfPOLnat[diffColumn]!=0]['Date'].to_list()
    values_of_change = dfPOLnat.dropna()[dfPOLnat[diffColumn]!=0][diffColumn].to_list()
    
    for index in range(len(dates_of_change)):
        DATELIST.append({
            'policy_name' : policyName,
            'dates_of_change':dates_of_change[index],
            'value' : values_of_change[index],
            'jur' : 'NAT'
        })
```

```python
NATPOLCHANGE = pd.DataFrame(DATELIST)
```

```python
NATPOLCHANGE
```

### State Policy

```python
# Policy Data 
dfPOLstate = dfPOL[dfPOL['Jurisdiction']=='STATE_TOTAL']
```

```python
dfPOLstate['RegionCode'] = dfPOLstate['RegionCode'].apply(lambda x: x.split('_')[1])
```

```python
STATEOP = []
```

```python

for stateName in dfPOLstate['RegionCode'].unique():
    stateDF = dfPOLstate[dfPOLstate['RegionCode']==stateName]
    polColumns = [
    'C1_School closing',
    'C6_Stay at home requirements',
    'C2_Workplace closing',
    'C7_Restrictions on internal movement',
    'C3_Cancel public events',
    'C4_Restrictions on gatherings',
    'C5_Close public transport',
    'H2_Testing policy'
    ]
    
    stateDF = stateDF[polColumns+['Date']]
    stateDF = stateDF.sort_values(by=['Date'])
    for policyName  in polColumns:
        stateDF['diff_'+policyName] = stateDF.loc[:,policyName]\
        .rolling(window=2)\
        .apply(lambda x: x.iloc[1] - x.iloc[0])
        
    DATELIST = []

    for policyName  in polColumns:
        diffColumn = 'diff_'+policyName

        dates_of_change = stateDF.dropna()[stateDF[diffColumn]!=0]['Date'].to_list()
        values_of_change = stateDF.dropna()[stateDF[diffColumn]!=0][diffColumn].to_list()

        for index in range(len(dates_of_change)):
            DATELIST.append({
                'policy_name' : policyName,
                'dates_of_change':dates_of_change[index],
                'value' : values_of_change[index],
                'jur' : stateName
            })
            
    STATEOP.append(pd.DataFrame(DATELIST))
    
```

```python
STATEPOLCHANGE = pd.concat(STATEOP)
```
